```python
# Part 2: Delving into Product Pricing
# Business Question: How are products priced on Amazon UK, and are there specific price points 
#     or ranges that are more common?

# Measures of Centrality:

# Calculate the mean, median, and mode for the price of products.
# What's the average price point of products listed? 
#    average price point = 89.24
# How does this compare with the most common price point (mode)?

# the mean/average is more sensitive to extreme values/outliers, here it's significantly higher
# than the most common price point, which indicates that there are products more expensive showing as outliers

mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

round(mean_price, 2), median_price, mode_price
```

```python
# Measures of Dispersion:

# Determine the variance, standard deviation, range, and interquartile range for product price.
# How varied are the product prices? Are there any indicators of a significant spread in prices?

variance_price = df['price'].var()
std_dev_price = df['price'].std()
min_price = df['price'].min()
max_price = df['price'].max()
range_price = max_price - min_price
quantiles_price = df['price'].quantile([0.25, 0.5, 0.75])

round(variance_price,2), round(std_dev_price,2), min_price, max_price, range_price, quantiles_price
```

```python
# Part 3: Unpacking Product Ratings
# Business Question: How do customers rate products on Amazon UK, and are there any patterns 
# or tendencies in the ratings?
```

```python
# Measures of Centrality:
df_w_zero = df[df['stars'] != 0] # this is creating a new DF filtering out the 0 values from the stars column

mean_stars = df_w_zero['stars'].mean()
median_stars = df_w_zero['stars'].median()
mode_stars = df_w_zero['stars'].mode()[0]

print('Mean:', round(mean_stars, 2))
print('Median:', median_stars)
print('Mode:', mode_stars)

print('\nHow do customers generally rate products? Is there a common trend? the common trend is 4.5 stars')
```

```python
# Measures of Dispersion:

# Determine the variance, standard deviation, and interquartile range for product rating.

variance_rate = df_w_zero['stars'].var()
std_dev_rate = df_w_zero['stars'].std()
min_rate = df_w_zero['stars'].min()
max_rate = df_w_zero['stars'].max()
range_rate = max_rate - min_rate
quantiles_rate = df_w_zero['stars'].quantile([0.25, 0.5, 0.75])

print('variance:' ,round(variance_rate,2))
print('std_dev_rate:',round(std_dev_rate,2))
print('min_rate:', min_rate)
print('max_rate:', max_rate)
print('range:', range_rate)
print('\nquantiles:\n',quantiles_rate)

print('\nAre the ratings consistent, or is there a wide variation in customer feedback?\n variation is normal, given that rates are consistent')
```

```python
# Shape of the Distribution:

# Calculate the skewness and kurtosis for the rating column.
# Are the ratings normally distributed, or do they lean towards higher or lower values?

skewness_rate = df_w_zero['stars'].skew()
kurtosis_rate = df_w_zero['stars'].kurtosis()

print('skewness_rate:',round(skewness_rate,2)) # this rate indicates that most products are highly rated with some outliers badly rated
print('kurtosis_rate:', round(kurtosis_rate,2)) # ratings lean towards higher values 
```

```python
# Visualizations:

# Plot a histogram to visualize the distribution of product ratings. Is there a specific rating 
# that is more common?

sns.histplot(df_w_zero['stars'], kde=True, bins=20, color="purple")
print('the most common rating is 4.5')
```
